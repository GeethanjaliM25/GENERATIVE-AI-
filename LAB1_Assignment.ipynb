```python
from google import genai
from getpass import getpass

api_key = getpass("Enter Gemini API Key: ")

client = genai.Client(api_key=api_key)
```

```python
print("SMART STUDY NOTES GENERATOR")
print("-"*40)

subject = input("Enter Subject: ")
topic = input("Enter Topic: ")
difficulty = input("Difficulty Level (Easy/Medium/Hard): ")
length = input("Length (Short/Medium/Detailed): ")
format_type = input("Format (Bullet Points / Explanation): ")
```

```python
prompt = f"""
You are an AI study assistant.

Generate study notes for the following topic.

Subject: {subject}
Topic: {topic}
Difficulty Level: {difficulty}
Length: {length}
Format: {format_type}

Instructions:
1. Make notes easy to understand
2. Use clear academic language
3. Avoid unnecessary information
4. Make it helpful for exam preparation
"""
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print("\nGENERATED NOTES")
print("-"*40)
print(response.text)
```
